```python
# Negation words handling for stopwords removal function

def stopwords_neg_handling():
  # Get list of stopwords from nltk
  sw = list(stopwords.words('indonesian'))

  # Make a list of negation words
  neg_words = negation_word = ['tidak', 'usah', 'mesti', 'jangan', 'bukan', 'enggak', 'tak', 'tanpa', 'ngga', 'ga', 'tempat', 'baik', 'lebih', 'sekitar', 'kira', 'kira-kira']

  # Make a new list to handle the negation words
  fix_stopwords = []

  i = 0
  j = 0
  for word in sw:
    if word not in neg_words:
      fix_stopwords.append(word)
      i = i + 1
    else:
      j = j + 1

  return fix_stopwords
```

```python
stopwords_fix = stopwords_neg_handling()
```

```python
# Stopwords removal function

def remove_stopwords(text, stopwords):

  return " ".join([word for word in str(text).split() if word not in stopwords])
```
